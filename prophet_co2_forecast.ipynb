```python
def prophet(country, year):
    df_con = df_a.loc[df_a['country'] == country]
    df_con.dropna(inplace=True)
    df_con_last = df_con.loc[df_con['year'] > year]
    df_con_fb = df_con_last.drop('country', axis=1)
    df_fbp = df_con_fb.rename(columns={'co2': 'y', 'year': 'ds'})
    df_fbp.plot(x='ds', y='y', figsize=(12, 8))
    train = df_fbp.iloc[:len(df_fbp) - int((0.1 * len(df_fbp)))]
    test = df_fbp.iloc[len(df_fbp) - int((0.1 * len(df_fbp))):]
    m = Prophet()
    m.fit(train)
    future = m.make_future_dataframe(periods=int((0.1 * len(df_fbp))), freq='Y')
    forecast = m.predict(future)
    plotpred = m.plot(forecast)
    predictions = forecast.iloc[-int((0.1 * len(df_fbp))):]['yhat']
    print('mean squared error is', mse(test['y'], predictions))
    print('root mean squared error is ', rmse(test['y'], predictions))
```

```python
def prophet_forecast(country, year, num_periods):
    df_con = df_a.loc[df_a['country'] == country]
    df_con.dropna(inplace=True)
    df_con_last = df_con.loc[df_con['year'] > year]
    df_con_fb = df_con_last.drop('country', axis=1)
    df_fbp = df_con_fb.rename(columns={'co2': 'y', 'year': 'ds'})
    df_fbp.plot(x='ds', y='y', figsize=(12, 8))
    m = Prophet()
    m.fit(df_fbp)
    future = m.make_future_dataframe(periods=num_periods, freq='Y')
    forecast = m.predict(future)
    plotpred = m.plot(forecast)
    return forecast
```
